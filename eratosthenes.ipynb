```julia
using QuadGK
using Plots
```

```julia
function eratosthenes( n )
    prims = [ true for i in 1:n ]
    
    sqrn = convert( Int64, floor(n^(1/2)))
    
    for i in 2:sqrn
        if prims[i] == true
            st = i^2
            while st <= n
                prims[st] = false
                st = st + i
            end
        end
    end
    
    return prims
end
```

```julia
k = 1_000_0000
prims = eratosthenes( k )
prims
```

```julia
numprimes = [0]
for i in 2:k
    if prims[i] 
        push!( numprimes, numprimes[end] + 1)
    else
        push!( numprimes, numprimes[end] )
    end
end
```

```julia
numprimes
```

```julia
numprimes[1_000_0000]
```

```julia
π(n) = numprimes[n]
λ(n) = n/log(n)
∫(n) = quadgk(t -> 1/log(t),2,n)[1]
```
